```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('veriler.csv')
```

```python
kazancToplama = data.groupby('Kişi')['Kazanılan Para'].sum().reset_index()
```

```python
toplamKazanc = kazancToplama['Kazanılan Para'].sum()
```

```python
kazancToplama['Yüzde'] = kazancToplama['Kazanılan Para'] / toplamKazanc * 100
```

```python
plt.pie(kazancToplama['Yüzde'], labels=kazancToplama['Kişi'], autopct='%1.1f%%')
```

```python
plt.title('Kazanılan Para Dağılımı')
```

```python
plt.axis('equal')
```

```python
for index, row in kazancToplama.iterrows():
    print(f"{row['Kişi']} - %{row['Yüzde']:.2f} - {row['Kazanılan Para']:.2f} lira")
```
